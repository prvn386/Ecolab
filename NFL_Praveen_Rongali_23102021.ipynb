## Mean Encoding is used as there many nominal based categories

```python
# missing value percentage in categorical columns
round((train_df[cat_col].isnull().sum()/len(train_df[cat_col]))*100,2).sort_values(ascending=False)
```

```python
# missing value percentage in numerical columns
round((df[num_col].isnull().sum()/len(df[num_col]))*100,2).sort_values(ascending=False)
```

```python
cat_col.remove('Player')
```

### Encoding of Categorical columns

```python
def encoding(columns,df):
    for x in columns:
        Mean_encoded_1 = df.groupby([x])['G'].mean().to_dict()
        df[x] =  round(df[x].map(Mean_encoded_1),2)
```

```python
encoding(cat_col,train_df)
```

### Model Buidling

```python
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

```python
scaler = MinMaxScaler()

train_df = scaler.fit_transform(df_n)
train_df = pd.DataFrame(train_df,columns = df_n.columns)
train_df.head()
```

```python
y_train = train_df.pop('G')
X_train = train_df
```

```python
lm = LinearRegression()

# Fitting a line
lm.fit(X_train, y_train)
```

```python
# Import RFE
from sklearn.feature_selection import RFE

# RFE with 14 features
lm = LinearRegression()
rfe14 = RFE(lm, 14)

rfe14.fit(X_train, y_train)

#result
print(rfe14.support_)           
print(rfe14.ranking_)  
```

```python
# columns which are top contributors for the model
cols = X_train.columns[rfe14.support_]
cols
```

```python
# Importing statsmodels
import statsmodels.api as sm  
# creating the dataframe using the columns selected by RFE
x_train_rfe14 = X_train[cols]

#add the column for constants
x_train_rfe14 = sm.add_constant(x_train_rfe14)
lm = sm.OLS(y_train, x_train_rfe14).fit()

#print the summary of the model
lm.summary()
```

```python
#Dropping College due to high P value
cols = cols.drop("College")

#add the column for constan
lm = sm.OLS(y_train, x_train_rfe14[cols]).fit()

#print the summary of the model
lm.summary()
```

```python
#Dropping FantPos due to high P value

cols = cols.drop("FantPos")

#add the column for constan
lm = sm.OLS(y_train, x_train_rfe14[cols]).fit()

#print the summary of the model
lm.summary()
```

```python
#Dropping Wonderlic due to high P value

cols = cols.drop("Wonderlic")

#add the column for constan
lm = sm.OLS(y_train, x_train_rfe14[cols]).fit()

#print the summary of the model
lm.summary()
```

```python
#Dropping W due to high P value

cols = cols.drop("W")

#add the column for constan
lm = sm.OLS(y_train, x_train_rfe14[cols]).fit()

#print the summary of the model
lm.summary()
```

```python
#Dropping Draft Round due to high P value

cols = cols.drop("Draft_Round")

#add the column for constan
lm = sm.OLS(y_train, x_train_rfe14[cols]).fit()

#print the summary of the model
lm.summary()
```

**Hometown,Home_State,YA,YR,DOB,Year,Cmp,Att2,Rec all these 9 features are the top independent
<br>features contributing for the model prediction**

```python
# prediction of y_train
y_train_pred = lm.predict(x_train_rfe14[cols])
```

```python
#plot the histogram for the residuals
fig = plt.figure()
sns.distplot((y_train-y_train_pred))
plt.xlabel("Error")
```

**Error value follows a normal distribution curve**
